## 1. Создайте мешок слов с помощью
sklearn.feature_extraction.text.CountVectorizer.fit_transform(). Применим его к 'tweet_stemmed'
и 'tweet_lemmatized' отдельно.

● Игнорируем слова, частота которых в документе строго превышает порог 0.9 с
помощью max_df.

● Ограничим количество слов, попадающий в мешок, с помощью max_features =
1000.

● Исключим стоп-слова с помощью stop_words='english'.

● Отобразим Bag-of-Words модель как DataFrame. columns необходимо извлечь с
помощью CountVectorizer.get_feature_names().


```python
from sklearn.feature_extraction.text import CountVectorizer
```

Для tweet_stemmed

```python
df['tweet_stemmed'] = df.tweet_stemmed.apply(lambda row: ' '.join(word for word in row))
  
# Создание мешка слов для 'tweet_stemmed' с помощью CountVectorizer

vectorizer_stemmed = CountVectorizer(stop_words='english', max_df=0.9, max_features=1000)
bag_of_words_stemmed = vectorizer_stemmed.fit_transform(df['tweet_stemmed'])

# Отобразим Bag-of-Words модель как DataFrame
feature_names = vectorizer_stemmed.get_feature_names()
pd.DataFrame(bag_of_words_stemmed.toarray(), columns = feature_names)
```

Для 'tweet_lemmatized'

```python
df['tweet_lemmatized'] = df.tweet_lemmatized.apply(lambda row: ' '.join(word for word in row))
  
# Создание мешка слов для 'tweet_lemmatized' с помощью CountVectorizer

vectorizer_lemmatized = CountVectorizer(stop_words='english', max_df=0.9, max_features=1000)
bag_of_words_lemmatized = vectorizer_lemmatized.fit_transform(df['tweet_lemmatized'])

# Отобразим Bag-of-Words модель как DataFrame
feature_names = vectorizer_lemmatized.get_feature_names()
pd.DataFrame(bag_of_words_lemmatized.toarray(), columns = feature_names)
```

## 3. Проверьте ваши векторайзеры на корпусе который использовали на вебинаре, составьте
таблицу метод векторизации и скор который вы получили (в методах векторизации по

```python
# Загружаем данные
data = open('./corpus').read()
labels, texts = [], []
for i, line in enumerate(data.split("\n")):
    content = line.split()
    labels.append(content[0])
    texts.append(" ".join(content[1:]))

# создаем df
trainDF = pd.DataFrame()
trainDF['text'] = texts
trainDF['label'] = labels
trainDF.head(5)
```

```python
from sklearn import model_selection, preprocessing, linear_model
# from sklearn.metrics import accuracy_score
from sklearn.metrics import *
```

```python
train_x, valid_x, train_y, valid_y = model_selection.train_test_split(trainDF['text'], trainDF['label'])

# labelEncode целевую переменную
encoder = preprocessing.LabelEncoder()
train_y = encoder.fit_transform(train_y)
valid_y = encoder.fit_transform(valid_y)  
```

```python
results_df = pd.DataFrame(columns=['name_of_vectorizer_with_parameters', 'score'])
results_df
```

```python
vectorizers = [
               vectorizer_stemmed,
               vectorizer_lemmatized,
               tfidf_vectorizer_lemmatized,
               tfidf_vectorizer_stemmed
               ]

for v in vectorizers:
  v.fit(trainDF['text'])

  xtrain_count =  v.transform(train_x)
  xvalid_count =  v.transform(valid_x)

  classifier = linear_model.LogisticRegression()
  classifier.fit(xtrain_count, train_y)
  predictions = classifier.predict(xvalid_count)

  acc_score = accuracy_score(valid_y, predictions)
  results_df = results_df.append({'name_of_vectorizer_with_parameters': v, 
                   'score': acc_score
                   }, ignore_index=True)
```

```python
from sklearn.decomposition import TruncatedSVD
```

```python
values = tfidf_vectorizer.fit_transform(trainDF['text'])
values.shape
```

```python
# CountVectorizer
res = pd.DataFrame()

for n_components in [1000, 3000, 5000]:
    svd = TruncatedSVD(n_components=n_components)
    
    xtrain_count =  count_vectorizer.transform(train_x)
    xvalid_count =  count_vectorizer.transform(valid_x)

    xtrain_count_pca =  svd.fit_transform(xtrain_count)
    xvalid_count_pca =  svd.transform(xvalid_count)

    classifier = linear_model.LogisticRegression()
    classifier.fit(xtrain_count_pca, train_y)
    
    predictions = classifier.predict(xvalid_count_pca)
    
    res = res.append(
        pd.DataFrame([[str(n_components), accuracy_score(valid_y, predictions)]], columns=['# Components', 'Score'])
    )
res.sort_values('Score', ascending=False, inplace=True)
res.head()
```

```python
#TFiDf 
res = pd.DataFrame()

for n_components in [1000, 3000, 5000]:
    svd = TruncatedSVD(n_components=n_components)
    
    xtrain_count =  tfidf_vectorizer.transform(train_x)
    xvalid_count =  tfidf_vectorizer.transform(valid_x)

    xtrain_count_pca =  svd.fit_transform(xtrain_count)
    xvalid_count_pca =  svd.transform(xvalid_count)

    classifier = linear_model.LogisticRegression()
    classifier.fit(xtrain_count_pca, train_y)
    
    predictions = classifier.predict(xvalid_count_pca)
    
    res = res.append(
        pd.DataFrame([[str(n_components), accuracy_score(valid_y, predictions)]], columns=['# Components', 'Score'])
    )
res.sort_values('Score', ascending=False, inplace=True)
res.head()
```
